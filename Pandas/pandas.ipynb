regex is regular expressions. Below are some powerful use cases


```python
# To find athletes born in cities that start with a vowel
vowel_cities = bios[bios['born_city'].str.contains(r'^[AEIOUaeiou],', na = False)]
vowel_cities.head()
```

```python
# to find athletes with names that contain exactly two vowels\
two_vowels = bios[bios['name'].str.contains(r'^[^AEIOUaeiou]*[AEIOUaeiou][AEIOUaeiou]*[^AEIOUaeiou]*$', na = False)]
two_vowels.head()
```

```python
# To find names that have repeated consecutive letters
repeated_letters = bios[bios['name'].str.contains(r'(.)\1', na = False)]
repeated_letters.head()
```

```python
# To find athletes with names ending in 'son' or 'sen'
son_sen_names = bios[bios['name'].str.contains(r'son$|sen$', case = False, na = False)]
son_sen_names.head()
```

```python
# TO find athletes born in year starting with '19'
born_19xx = bios[bios['born_date'].str.contains(r'^19', na = False)] .head()
born_19xx
```

na = False is just a way t handle null values in the row.By making it false , it makes contains always false.

TO disable these regex features, regex = False

```python
import sqlite3
conn = sqlite3.connect('example.db')
cur  = conn.cursor()
cur.execute("""
  CREATE TABLE IF NOT EXISTS sales (
    order_id INTEGER PRIMARY KEY,
    customer TEXT,
    amount REAL,
    order_date TEXT
  )
""")
conn.commit()
cur.executemany(
  "INSERT OR IGNORE INTO sales VALUES (?,?,?,?)",
  [(1,'Alice',125.5,'2025-01-15'),
   (2,'Bob',230.0,'2025-02-20')]
)
conn.commit()
```

```python

df_sql = pd.read_sql_query(
    "SELECT * FROM sales",
    conn,
    index_col="order_id"
)
print(df_sql)
```

```python
from sqlalchemy import create_engine

# Create a SQLAlchemy engine (instead of sqlite3.connect)
engine = create_engine("sqlite:///example.db")

# Now you can load the entire table by name
df_sql = pd.read_sql_table(
    "sales",
    con=engine,
    index_col="order_id"
)
print(df_sql)
```

read_sql_query vs. read_sql vs. read_sql_table:

    read_sql_query(sql, con) executes arbitrary SQL (e.g. SELECT * FROM sales).

    read_sql(sql, con) is a convenience wrapper that dispatches to query or table based on its first argument—but with a raw DBAPI connection, pandas treats the argument as SQL, so it must be valid SQL.

    read_sql_table(table_name, con) requires a SQLAlchemy engine; it inspects the database schema and generates a SELECT * automatically.

DBAPI vs. SQLAlchemy connections:

    DBAPI (sqlite3.connect) is built into Python, simple but limited (only SQLite).

    SQLAlchemy engine supports many databases uniformly, enables introspection (needed for read_sql_table), connection pooling, and more flexible URI strings.

index_col usage:
Converting a unique column (like order_id) into the DataFrame index lets you leverage fast lookup, alignment, and join operations in pandas.

Always write valid SQL when using read_sql_query to avoid syntax errors.

For large tables, consider streaming results with chunksize to control memory usage.

Specifying column types or using SQL-side filters reduces data transferred into pandas, improving speed.
