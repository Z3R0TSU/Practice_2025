```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# plt.plot(x,y, label = '2x', color = 'red', linewidth = 2, marker = '*' , linestyle = '--', markersize = 10, markeredgecolor = 'blue') # color can be hex-code as well
# To use shorthand notation of the above line
plt.plot(x,y, 'b^--' ,label = '2x')

# second line


plt.title("Graph", fontdict={'fontname' : 'Arial' , 'fontsize' : 20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
# to choose the x axis and y axis scale
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

plt.legend()
```
