```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# To resize the graph
plt.figure(figsize=(5,3), dpi = 350)

# Line 1
# Keyword Argument Notation
# plt.plot(x,y, label = '2x', color = 'red', linewidth = 2, marker = '*' , linestyle = '--', markersize = 10, markeredgecolor = 'blue') # color can be hex-code as well
# To use shorthand notation
plt.plot(x,y, 'b^--' ,label = '2x')

# Line 2

# select interval we want to point plots at start - end - interval
x2 = np.arange(0,4.5,0.5)
#plt.plot(x2,x2**2, 'r', label = 'X^2')
# to show projection 

# to plot upto 5 as line graph
plt.plot(x2[:6],x2[:6]**2, 'r', label = 'X^2')
# to ploy from 6 (inclusive) as dotted
plt.plot(x2[5:],x2[5:]**2, 'g--')

# for title
plt.title("Graph", fontdict={'fontname' : 'Arial' , 'fontsize' : 20})
# for x and y axis
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# to choose the x axis and y axis scale
plt.xticks([0,1,2,3,4])
#plt.yticks([0,2,4,6,8,10])

# add a legend to the graph
plt.legend()

# To save the graph
plt.savefig('mygraph.png' , dpi = 250)
```

plt.show() renders all the figures built so far. But for the compiler to show the output it should be of inline display. This can be manually be enabled by using %matplotlib inline

Bar Chart

```python
labels = ['A' , 'B' , 'C']
values = [1,4,2]

bars = plt.bar(labels,values)

patterns = ['/' , 'o' , '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0)) # if it is pop() then it would take the last first so the first bar will have stars

# same as above
#bars[0].set_hatch('/')
#bars[1].set_batch('o')
#bars[2].set_batch('*')

plt.figure(figsize=(6,4))
```
