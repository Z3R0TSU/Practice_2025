```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
gas = pd.read_csv('./data/gas_prices.csv')
gas.head()
```

```python
plt.title('Gas Prices over time ($USD)')

plt.plot(gas.Year,gas.USA, 'b.-', label = 'USA') # cannot add color and line type argument after label
plt.plot(gas['Year'],gas['Canada'], 'g*-', label = 'Canada') # gas.South Korea wouldnt work as there is a space. so we would have to use gas['South Korea']
plt.plot(gas.Year,gas['South Korea'], 'r^-', label = 'South Korea')
plt.plot(gas.Year,gas['Australia'], 'y^-', label = 'Australia')


plt.xticks(gas.Year[::3])
plt.legend()

plt.xlabel('Years')
plt.ylabel('Price ($)')
```

```python
# another way to plot the above graph
plt.title('Gas Prices over time ($USD)')


countries_of_interest = ['Australia', 'USA', 'Canada', 'South Korea']
for country in gas:
    if country in countries_of_interest:
        plt.plot(gas['Year'], gas[country], marker = '.', label = country)

plt.xticks(gas['Year'][::3])
plt.xlabel('Years')
plt.ylabel('Price ($)')
plt.legend(loc='best')
plt.show()
```

```python
# we can add font dict to any text based labels
plt.title('Gas Prices over time ($USD)', fontdict={'fontweight':'bold', 'fontsize' : 18})

for country in gas.columns:
    if country == 'Year':
        continue
    # use the variable `country`, not the literal string 'country'
    plt.plot(gas['Year'], gas[country], marker='.', label=country)

plt.xticks(gas['Year'][::3].tolist()+[2011]) # we cant directly add 2011 as it needts to be a list so we can add 2011 as list , else 2011 will be numerically added to all years which is undesired
plt.xlabel('Years')
plt.ylabel('Price ($)')
plt.legend(loc='best')
plt.savefig("Gas_Price_Fig.png", dpi = 300)
plt.show()
```
